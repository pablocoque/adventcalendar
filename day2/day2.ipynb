```python
class InvalidIDs():
    def __init__(self, filename):
        self.filename = filename
        self.ranges = []
        self.first_invalid_ids = []
        self.second_invalid_ids = []

    def read_ranges(self):
        with open(self.filename, 'r') as file:
            data = file.read()
            ranges = data.split(',')
            for r in ranges:
                low, high = map(int, r.split('-'))
                self.ranges.append((low, high))
    
    def find_first_invalid_ids(self):
        for irange in self.ranges:
            for id in range(irange[0], irange[1] + 1):
                id_str = str(id)
                length = len(id_str)
                if id_str[0:length//2] == id_str[length//2:]:
                    self.first_invalid_ids.append(id)
    
    def find_second_invalid_ids(self):
        for irange in self.ranges:
            for id in range(irange[0], irange[1] + 1):
                id_str = str(id)
                length = len(id_str)
                for pattern_length in range(1, length//2 + 1):
                    invalid = True
                    for i in range(0, length - pattern_length, pattern_length):
                        if invalid:
                            if id_str[i:i+pattern_length] != id_str[i+pattern_length:i+2*pattern_length]:
                                invalid = False
                    if invalid:
                        self.second_invalid_ids.append(id)
                        break

    def get_sum_invalid_ids(self, which_set):
        sum = 0
        if which_set == "first":
            for id in self.first_invalid_ids:
                sum += id
            return sum
        elif which_set == "second":
            for id in self.second_invalid_ids:
                sum += id
            return sum
        else:
            raise ValueError("Invalid set specified.")
```

```python
test = InvalidIDs('test_input.txt')
test.read_ranges()
test.find_first_invalid_ids()
test.find_second_invalid_ids()
```

```python
test.get_sum_invalid_ids('first')
```

```python
test.get_sum_invalid_ids('second')
```
