```python
class EmergencyBattery():
    def __init__(self):
        self.banks = []
        self.maximum_joltage = []
        self.total_joltage = 0
    
    def load_batterybanks(self, banksfile):
        with open(banksfile, 'r') as f:
            for line in f:
                self.banks.append(line.strip())

    def calculate_max_bankjoltage(self, num_batteries = 2):
        for bank in self.banks:
            length = len(bank)
            max_batteries = [0]*num_batteries
            idx_of_prevmax = [-1]*(num_batteries + 1)
            for n in range(num_batteries):
                for i in range(idx_of_prevmax[n] + 1, length - num_batteries + n + 1):
                    if int(bank[i]) > max_batteries[n]:
                        max_batteries[n] = int(bank[i])
                        idx_of_prevmax[n + 1] = i

            joltage = 0
            for n in range(num_batteries):
                joltage += max_batteries[n] * (10 ** (num_batteries - n - 1))
            self.maximum_joltage.append(joltage)

    def calculate_total_joltage(self):
        self.total_joltage = sum(self.maximum_joltage)
        return self.total_joltage
```

```python
testbank = EmergencyBattery()
testbank.load_batterybanks('test_input.txt')
testbank.calculate_max_bankjoltage(num_batteries=12)
print(testbank.maximum_joltage)
testbank.calculate_total_joltage()
```
