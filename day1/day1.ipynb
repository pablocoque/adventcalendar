```python
class NorthPoleSafe():
    def __init__(self):
        self.min_position = 0
        self.max_position = 99
        self.times_zero = 0
        self.current_position = 50
        self.last_position = 0
        self.crossed_zero = 0
    
    def do_rotation(self, rotation):
        self.last_position = self.current_position
        decode = rotation.strip()
        sense = decode[0].upper()
        number_steps = int(decode[1:].strip())
        crossed_last = False
        if sense == 'L':
            sign = -1
        elif sense == 'R':
            sign = 1
        else:
            raise ValueError(f"Invalid rotation sense: {sense}")
        
        self.crossed_zero += number_steps // (self.max_position + 1)

        self.current_position += sign * (number_steps % (self.max_position + 1))

        if self.current_position < self.min_position:
            self.current_position += self.max_position + 1
            crossed_last = True
        if self.current_position > self.max_position:
            self.current_position -= self.max_position + 1
            crossed_last = True

        if self.current_position == self.min_position and crossed_last:
            crossed_last = False
        
        if crossed_last and self.last_position != self.min_position:
            self.crossed_zero += 1

        if self.current_position == self.min_position:
            self.times_zero += 1
            
        return

    def get_current_position(self):
        return self.current_position
    
    def get_times_zero(self):
        return self.times_zero

    def get_number_zero_crossings(self):
        return self.crossed_zero
    
    def get_Password(self):
        return self.crossed_zero + self.times_zero
```

```python
# Test this
safelock = NorthPoleSafe()
with open('test_input.txt', 'r') as f:
    print(f"Starting position: {safelock.get_current_position()}")
    for rotation in f:
        print(f"doing rotation: {rotation.strip()}")
        safelock.do_rotation(rotation.strip())
        print(f"times zero crossed: {safelock.get_number_zero_crossings()}")
        print(f"current position: {safelock.get_current_position()}")
print(f"times zero position: {safelock.get_times_zero()}")
```
